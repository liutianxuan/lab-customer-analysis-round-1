2. Show the DataFrame's shape.

```python
import pandas as pd
file1 = pd.read_csv('file1.csv')
file2 = pd.read_csv('file2.csv')
file3 = pd.read_csv('file3.csv')
display(file1.shape, file2.shape, file3.shape)
display(file1.columns, file2.columns, file3.columns)
```

3. Standardize header names.

4. Rearrange the columns in the dataframe as needed (Answer: see Q6)

```python
file2.columns = ['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']
file3.columns = ['Customer', 'State', 'Gender', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints',
       'Policy Type', 'Vehicle Class', 'Total Claim Amount']

file1 = file1.rename (columns = {'ST':'State'})
file2 = file2.rename (columns = {'ST':'State'})
file3 = file3.rename (columns = {'Gender':'GENDER'})

cols1 = []
for i in file1.columns:
    cols1.append(i.lower())
file1.columns = cols1

cols2 = []
for i in file2.columns:
    cols2.append(i.lower())
file2.columns = cols2

cols3 = []
for i in file3.columns:
    cols3.append(i.lower())
file3.columns = cols3

```

5. Correct the values in the column 'customer lifetime value'. They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)

```python
#display(file1.head(), file2.head(), file3.head())
file1['customer lifetime value']  = file1['customer lifetime value'].str.rstrip('%').astype('float')/100
file2['customer lifetime value']  = file2['customer lifetime value'].str.rstrip('%').astype('float')/100
#file1['customer lifetime value'] = pd.to_numeric(file1['customer lifetime value'], errors = 'coerce')
display(file1.head(), file2.head())
```

6. Concatenate the three dataframes

```python
customers = pd.concat([file1, file2, file3], axis=0)
customers.columns = customers.columns.str.replace(" ", "_")
customers = customers.reset_index(drop=True)
display(customers.head(), customers.tail())
print (customers.shape)
```
